##**1. Operaciones básicas**
**Objetivo:** Crear y manipular arrays.\
**Instrucciones:**
1. Crea un array de 10 elementos con valores entre 1 y 10.
2. Calcula la media, mediana, y desviación estándar de los valores.
3. Multiplica cada valor por 2.

### **Código y Resultados**

```python
# Creacion del array
import numpy as np

arr = np.random.randint(0, 11, 10)
print('arreglo: ', arr)

# Media
media = arr.mean()
print('media: ', media)

# mediana
mediana = np.median(arr)
print('mediana: ', mediana)

# Desviación estándar
desviacion = np.std(arr)
print(f"desviación estándar: {desviacion:.2f}")

# Multiplicación de los valores del arreglo * 2
arr_multiplicado = arr * 2
print('arreglo multiplicado: ', arr_multiplicado)
```

##**2. Operaciones con DataFrames**
**Objetivo:** Crear y manipular un DataFrame.\
**Instrucciones:**
1. Crea un DataFrame con 3 columnas (A, B, C) y 5 filas con números aleatorios enteros.
2. Renombra las columnas a Col1, Col2, Col3.
3. Añade una nueva columna que sea la suma de las columnas Col1 y Col2.
4. Elimina todas las filas donde el valor de Col3 sea menor que 10.

### **Código y Resultados**

```python
# Creación del DataFrame
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(0, 101, size=(5, 3))) # size=5 filas, 3 columnas
print(df,'\n')

# Renombrar las columnas
df = df.rename(columns={0: 'Col1', 1: 'Col2', 2: 'Col3'})
print('Columnas renombadas\n', df,'\n')

# Nueva columna
df['Col4'] = df['Col1'] + df['Col2']
print('Nueva columna\n', df,'\n')

# Eliminar filas
df = df[df['Col3'] >= 10]
print('Columnas después de la eliminación\n', df)
```

##**3. Estadística descriptiva**
**Objetivo:** Aplicar análisis estadístico básico.\
**Instrucciones:**
1. Crea un DataFrame con 100 valores aleatorios entre 0 y 50.
2. Calcula la media, mediana y moda de los valores.
3. Encuentra el valor mínimo y el valor máximo del DataFrame.

### **Código y Resultados**

```python
import pandas as pd
import numpy as np

# Creación de DataFrame
df = pd.DataFrame(np.random.randint(0, 51, size=(100, 1)))
print('DataFrame\n', df,'\n')

# Media
media = np.mean(df)
print('media: ', media)

# Mediana
mediana = np.median(df)
print('mediana: ', mediana)

# Moda
moda = df.mode()
print('moda: ', moda[0].to_string(index=False))

# Valor mínimo
minimo = np.min(df)
print('valor mínimo: ', minimo)

# Valor máximo
maximo = np.max(df)
print('valor máximo: ', maximo)
```

2. Genera 100 valores aleatorios asociados a estas fechas.

##**6. Gráfico de Barras**
**Objetivo:** Visualizar datos categóricos.\
**Instrucciones:**
1. Carga el dataset tips de Seaborn.
2. Realiza un gráfico de barras que muestre el total de las cuentas (total_bill)
agrupadas por día.
3. Añade etiquetas a los ejes y un título.

### **Código y Resultados**

```python
# Cargar biblioteca seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset tips de Seaborn
tips = sns.load_dataset('tips')

# Gráfico de barras
ax = sns.barplot(x='day', y='total_bill', data=tips, hue='day', errorbar=None)
for p in ax.patches:
    height = p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2,   # posición X: centro de la barra
        height + 1,                      # posición Y: un poquito arriba de la barra
        f'{float(height):.2f}',          # texto: total decimal
        ha='center',
        va='top'
    )
plt.title('Total de cuentas agrupadas por día')
plt.xlabel('Día')
plt.ylabel('Total de cuentas')
plt.show()
```

##**9. Heatmap**
**Objetivo:** Visualizar la correlación entre variables numéricas.\
**Instrucciones:**
1. Usa el dataset iris de Seaborn.
2. Calcula la matriz de correlación de las variables.
3. Crea un heatmap para visualizar la matriz de correlación.

### **Código y Resultados**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargando Dataset
iris = sns.load_dataset('iris')

# Solo las variables numéricas
iris_num = iris.select_dtypes(include=['float64', 'int64'])

# Matriz dew correlación
matriz_corr = iris_num.corr()

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de correlación entre variables numéricas')
plt.show()
```

##**10. Gráfico de Línea con Media Móvil**
**Objetivo:** Visualizar series temporales.
**Instrucciones:**
1. Genera una serie temporal de 100 días con valores aleatorios.
2. Calcula la media móvil de los últimos 10 días y grafícala junto a la serie original.
3. Usa diferentes colores para la serie original y la media móvil.

### **Código y Resultados**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creación de días
dias = pd.date_range(start='2024-01-01', periods=100, freq='D')

# Serie
np.random.seed(1)
valores = np.random.randn(100).cumsum() # Serie aleatoria acumulada para más variación
serie = pd.Series(valores, index=dias)

# Media móvil
media_movil = serie.rolling(window=10).mean()

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(serie.index, serie, label='Serie Original', color='blue')
plt.plot(media_movil.index, media_movil, label='Media móvil (últimos 10 días)', color='red')
plt.title('Serie temporal con media móvil')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.tight_layout()
plt.grid(True, linestyle='--')
plt.show()
```

##**11. Datos Temporales con Análisis Estadístico**
**Objetivo:** Trabajar con series temporales y calcular estadísticas.
**Instrucciones:**\
1. Genera una serie temporal de 200 días con valores aleatorios.
2. Calcula la media y la desviación estándar de la serie temporal.
3. Visualiza la serie junto con las bandas de desviación estándar (es decir, la media ± 1 desviación estándar).

### **Código y Resultados**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creación de días
dias = pd.date_range(start='2024-01-01', periods=200, freq='D')

# Valores
np.random.seed(2)
valores = np.random.randn(200)
serie = pd.Series(valores, index=dias)

# Media y desviación estándar
media = serie.mean()
desviacion_estandar = serie.std()

print(f'media: {media:.4f}')
print(f'desviación estándar: {desviacion_estandar:.4f}')

# Bandas de desviación estándar
banda_superior = media + desviacion_estandar
banda_inferior = media - desviacion_estandar

# Gráfica
plt.figure(figsize=(12, 6))
plt.fill_between(serie.index, banda_inferior, banda_superior, alpha=0.2,
                 label='Media ± 1 desviación estándar', color='blue')
plt.plot(serie.index, serie, label='Serie Temporal', color='maroon')
plt.axhline(media, linestyle='--', label='Media', color='navy')
plt.title('Serie Temporal con Bandas de Desviación Estándar')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.tight_layout()
plt.show()
```
