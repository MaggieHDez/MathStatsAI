```python
import math
import scipy.stats as stats
import random
import matplotlib.pyplot as plt

print("*********************EJERCICIO 1*********************")
# Datos para el ejercicio 1
n1 = 12
p1 = 0.10
q1 = 1 - p1

# a) Probabilidad de que más de 8 personas vean el programa: sumatoria para x=9,10,11,12
P_mas_de_8 = sum(math.comb(n1, x) * (p1**x) * (q1**(n1 - x)) for x in range(9, n1+1))
print("\na) Más de 8 personas P(X>8) =", P_mas_de_8)

# b) Probabilidad de que al menos una persona lo vea: 1 - P(X=0)
P_al_menos_uno = 1 - (q1)**n1
print("b) Al menos una persona  P(X>=1) = {:.5f} ó {:.3f} %".format(P_al_menos_uno, P_al_menos_uno*100))

# c) Media y Desviación Típica
mean1 = n1 * p1
std1 = (n1 * p1 * q1) ** 0.5
print("c) Media = {:.1f} y Desviación Típica = {:.5f}".format(mean1, std1))

print("\n*********************EJERCICIO 2*********************")
# Datos para el ejercicio 2
n2 = 20
p2 = 0.25
q2 = 1 - p2

# a) Probabilidad de contestar todas las preguntas mal (k = 0)
P_todo_mal = q2 ** n2
print("\na) Conteste todas las preguntas mal P(X=0) = {:.8f} ó {:.6f} %".format(P_todo_mal, P_todo_mal*100))

# b) Probabilidad de contestar al menos 10 preguntas bien: suma para k=10 a 20
P_al_menos_10 = 1 - sum(math.comb(n2, k) * (p2**k) * (q2**(n2 - k)) for k in range(0, 10))
print("b) Al menos 10 preguntas bien P(>=10) = {:.7f} ó {:.5f} %".format(P_al_menos_10, P_al_menos_10*100))

# c) Probabilidad de contestar entre 5 y 8 preguntas bien (k = 5, 6, 7, 8)
P_entre_5_y_8 = sum(math.comb(n2, k) * (p2**k) * (q2**(n2 - k)) for k in range(5, 9))
print("c) Conteste entre 5 y 8 preguntas bien. P(5≤X≤8) = {:.6f} ó {:.4f} %".format(P_entre_5_y_8, P_entre_5_y_8*100))

# d) Probabilidad de contestar menos de 3 preguntas bien (k = 0, 1, 2)
P_menos_de_3 = sum(math.comb(n2, k) * (p2**k) * (q2**(n2 - k)) for k in range(0, 3))
print("d)  Conteste menos de tres preguntas bien. P(X<3) = {:.7f} ó {:.5f} %".format(P_menos_de_3, P_menos_de_3*100))

print("\n*********************EJERCICIO 3*********************")

# Datos para el ejercicio 3:
N = 15      # Tamaño total de hombres y mujeres
n = 5       # Tamaño del comité

# Para el cálculo del comité de 2 hombres y 3 mujeres:
K_men = 5   # Total de hombres en N
k_hombres = 2   # Número de hombres en el comité
k_3_mujeres = 3  # Número de mujeres en el comité

# Para el comité solo por mujeres:
K_women = 10  # Total de mujeres en N
k_mujeres = 5  # Que las 5 personas del comité sean mujeres


# Probabilidad de tener 2 hombres y 3 mujeres:
p_2hombres_3mujeres = (math.comb(K_men, k_hombres) * math.comb(K_women, k_3_mujeres)) / math.comb(N, n)

# Probabilidad de tener solo mujeres (0 hombres, o equivalentemente 5 mujeres de 10):
p_solo_mujeres = math.comb(K_women, k_mujeres) / math.comb(N, n)

print(f"\na) Probabilidad de 2 hombres y 3 mujeres: {p_2hombres_3mujeres} ó {p_2hombres_3mujeres*100:.2f} %")
print(f"b) Probabilidad de solo mujeres: {p_solo_mujeres} ó {p_solo_mujeres*100:.2f} %")

# Datos para el ejercicio 4:
print("\n*********************EJERCICIO 4*********************")

# Definir los valores de a y r
a = 0.5
r = 0.5

# Usando la fórmula para la suma infinita de una serie geométrica
S_infinita = a / (1 - r)
print(f"\nSuma infinita (usando la fórmula) a / (1 - r) donde a = {a} y r = {r}: \n {a} / (1 - {r}) =  {a} / {1-r} = ", S_infinita)

# Datos para el ejercicio 5:
print("\n*********************EJERCICIO 5*********************")

# Número de términos (de tamaño grande para aproximar una suma infinita)
N = 100

# Para a)par:
# k = 2m donde m = 1, 2, 3, ... y
# P(e_{2m}) = (0.5)^(2m) = (0.25)^m
P_par = sum((0.25)**m for m in range(1, N+1))

# Para b)impar:
# k = 2m+1 donde m = 0, 1, 2, ... y
# P(e_{2m+1}) = (0.5)^(2m+1) = 0.5 * (0.25)^m
P_impar = sum(0.5 * (0.25)**m for m in range(0, N+1))

print("\nProbabilidad de que la primera 'águila' aparezca en un número PAR de volados:")
print("Aproximado: {} ó {:.2f} %".format(P_par, P_par*100))
print("Teórico: {} ó {:.2f} %".format(0.25/(1 - 0.25),(0.25/(1 - 0.25))*100))  # 0.25 / 0.75 = 1/3

print("\nProbabilidad de que la primera 'águila' aparezca en un número IMPAR de volados:")
print("Aproximado: {} ó {:.2f} %".format(P_impar, P_impar*100))
print("Teórico: {} ó {:.2f} %".format(0.5/(1 - 0.25), (0.5/(1 - 0.25))*100))   # 0.5 / 0.75 = 2/3

print("\n*********************EJERCICIO 7*********************")

# Número de simulaciones (lanzamientos de dos dados)
num_simulaciones = 1000000

# Contadores para los eventos de interés:
conteo_suma7 = 0
conteo_mayor10 = 0

# Simulación
for _ in range(num_simulaciones):
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    suma = dado1 + dado2
    if suma == 7:
        conteo_suma7 += 1
    if suma > 10:
        conteo_mayor10 += 1

# Cálculo de probabilidades experimentales
prob_suma7_exp = conteo_suma7 / num_simulaciones
prob_mayor10_exp = conteo_mayor10 / num_simulaciones

# Muestra diferentes resulados por la generación de los numeros aleatorios pero se acerca al porcentaje teórico
print("\nProbabilidad experimental de que la suma sea 7: {:.5f} ó {:.5f} %".format(prob_suma7_exp,prob_suma7_exp*100))
print("Probabilidad experimental de que la suma sea mayor que 10: {:.5f} ó {:.5f} %\n".format(prob_mayor10_exp,prob_mayor10_exp*100))

# Para visualizar, se puede graficar la distribución de las sumas
resultados = []
for _ in range(num_simulaciones):
    resultados.append(random.randint(1, 6) + random.randint(1, 6))

plt.hist(resultados, bins=range(2,14), edgecolor="black", density=True)
plt.xlabel("Suma")
plt.ylabel("Densidad")
plt.title("Histograma de la suma de dos dados")
plt.xticks(range(2,13))
plt.show()

print("\n*********************EJERCICIO 8*********************")

# Ejemplo: Obtener el valor z para el percentil 99
percentil = 0.99
z_valor = stats.norm.ppf(percentil).round(2)
print("\nEl valor z para el percentil 99 es:", z_valor)

# Para transformar este z a una escala con media (mu) y desviación estándar (sigma) específicas:
mu = 120
sigma = 6
x_valor = mu + z_valor * sigma
print("El IQ correspondiente (usando mu=120 y sigma=6) es:", x_valor)

# Ejemplo: Obtener el valor z para el percentil 1 (para determinar un umbral inferior)
percentil_inferior = 0.01
z_valor_inferior = stats.norm.ppf(percentil_inferior).round(2)
print("El valor z para el percentil 1 es:", z_valor_inferior)

# Transformamos a la escala deseada:
x_valor_inferior = mu + z_valor_inferior * sigma
print("El IQ correspondiente al percentil 1 es:", x_valor_inferior)

# Probabilidad de que el IQ esté entre 115 y 120 puntos
# Convertir los valores a la escala z:
z_115 = (115 - mu) / sigma
z_120 = (120 - mu) / sigma

# Calcular las probabilidades acumuladas
prob_115_120 = stats.norm.cdf(z_120) - stats.norm.cdf(z_115)
# Mostrar la probabilidad como número decimal y porcentaje
print("Probabilidad de que el IQ esté entre 115 y 120: {:.4f} ó {:.2%}".format(prob_115_120, prob_115_120))

# Probabilidad de que el IQ esté entre 125 y 130 puntos
# Convertir los valores a la escala z:
z_125 = (125 - mu) / sigma
z_130 = (130 - mu) / sigma

# Calcular la probabilidad del intervalo
prob_125_130 = stats.norm.cdf(z_130) - stats.norm.cdf(z_125)
# Mostrar la probabilidad como número decimal y porcentaje
print("Probabilidad de que el IQ esté entre 125 y 130: {:.4f} ó {:.2%}".format(prob_125_130, prob_125_130))

print("\n*********************EJERCICIO 9*********************")

# Probabilidades de éxito en cada experimento
p_A = 1 - (5/6)**4        # Aproximadamente 0.5177
p_B = 1 - (35/36)**24      # Aproximadamente 0.4920

# Diferencia de probabilidades
delta_p = p_A - p_B       # Aproximadamente 0.0257

# Varianzas individuales y suma de varianzas (para una sola repetición)
var_A = p_A * (1 - p_A)
var_B = p_B * (1 - p_B)
variance_total = var_A + var_B   # Aproximadamente 0.499

def experimentos_necesarios(z_crit, delta_p, variance_total):
    n = variance_total * (z_crit / delta_p) ** 2
    return n

# Valores críticos para la prueba unilateral
z_crit_70 = 0.524     # Para 70% de certidumbre
z_crit_95 = 1.645     # Para 95% de certidumbre

# Calcular el número de experimentos necesarios
n_70 = experimentos_necesarios(z_crit_70, delta_p, variance_total)
n_95 = experimentos_necesarios(z_crit_95, delta_p, variance_total)

print("\nNúmero de experimentos necesarios para 70% de certidumbre: {:.0f}".format(n_70))
print("Número de experimentos necesarios para 95% de certidumbre: {:.0f}".format(n_95))
```
